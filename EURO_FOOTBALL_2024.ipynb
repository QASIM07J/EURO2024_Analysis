# Data Analysis


```python
sns.barplot(x=Age_Count.index, y=Age_Count.values)


plt.xticks(rotation=45)
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.title('Distribution of Ages in the Sample Data')
plt.show()
```

```python
#Similarly we check how many players are participating from each country
country_counts = df['Country'].value_counts()

plt.figure(figsize=(8, 6))
ax = country_counts.plot(kind='bar')

# Add labels to each bar
for i, v in enumerate(country_counts):
    plt.text(i, v + 0.1, str(v), ha='center', va='bottom')

plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.title('Players by Country')
plt.show()
```

```python
df
```

```python
#Lets try to check which club players are participating the most.

club_counts = df['Club'].value_counts().head(10)

club_counts.plot(kind = 'bar' , title='Most Club Players')
plt.xlabel("Club")
plt.ylabel("Count")
plt.show()
```

```python
# Find the player with the highest market value

max_value = df['MarketValue'].max()
player_with_max_value = df[df['MarketValue'] == max_value]

print("Player with the highest market value:")
print(player_with_max_value[['Name', 'MarketValue']])
```

```python
top_10_players = df.sort_values(by='MarketValue', ascending=False).head(10)

# Print the top 10 players with their market values
print("Top 10 Players with Highest Market Value:")
print(top_10_players[['Name', 'MarketValue']])
```

```python
# Create the line graph
plt.figure(figsize=(10, 6))
plt.plot(top_10_players['Name'], top_10_players['MarketValue'], marker='o')
plt.title('Top 10 Players by Market Value')
plt.xlabel('Player Name')
plt.ylabel('Market Value')
plt.xticks(rotation=45) 
# plt.grid(True) You can uncheck for better visibility 
plt.show()


#If you want to replace the float to orignal marketvalue then first convert it to string using the command
#df['MarketValue'] = df['MarketValue'].astype(str)
```
